# Notebook 3:  Redes de Kohonen y SOM (Self organized maps)
En este notebook vamos a implementar el algoritmo de clustering SOM en R. El objetivo es entender como funciona este algoritmo y  aprender a utilizarlo con un ejemplo simple. Recuerde instalar previamente el paquete kohonen en R.



Nota: 4.0

```r
library(kohonen)
set.seed(7)

#ruta datos
#lee datos del archivo de texto como una matriz. La primera #columna se utiliza como vector de nombres de las filas
datos <-read.table("milk1.dat",row.names=1,dec=".",header=TRUE)
datos <- as.data.frame(datos)
summary(datos)
```

```r
#create SOM grid
sommap <- som(scale(datos), grid = somgrid(2, 2, "hexagonal"))
plot(sommap)

colors <- function(n, alpha = 1) {
    rev(heat.colors(n, alpha))
}

plot(sommap, type = "counts", palette.name = colors, heatkey = TRUE)

plot(sommap, type = "mapping", pchs = 20, main = "Mapping Type SOM")

plot(sommap, type = "dist.neighbours", palette.name = terrain.colors)
```

```r
mat <- matrix(sommap$codes[[1]], ncol=4,nrow=4,byrow=FALSE)

## use hierarchical clustering to cluster the codebook vectors
groups<-3
som.hc <- cutree(hclust(dist(mat)), groups)

#plot
plot(sommap, type="codes", bgcol=rainbow(groups)[som.hc])

#cluster boundaries
add.cluster.boundaries(sommap, som.hc)
```

### a) Explique como funciona el método som, describa parámetros de entrada y salida y los gráficos obtenidos.



* Este método consiste en una grilla de cuadrados o hexágonos, donde se almacenan datos similares y además las grillas colindantes contienen datos con cierto grado de similitud.

* El algoritmo se compone de las siguientes etapas:

**_Etapa 1:_** 

1) Definir las grillas

2) Inicializar los pesos $(w)$ con valores al azar y seleccionar un tamaño de vecinandario   inicial ,tal que cubra la mitad de las celdas.

3) Seleccionar un registro al azar $x_i$  del dataset.

4) Encontrar el nodo k que tenga el vector de pesos $w_k$ más cercano a x. Para esto se puede la distancia Euclídea (d). $$d(x−w_k)=min_i[d(x−w_i)]$$

**_Etapa 2:_**

5) Actualizar los pesos de los nodos en la vecindad de k utilizando la regla de aprendizaje de Kohonen:

$$\mbox{Si i}  \in N_k→ w^{new}_i=w^{old}_i+ \infty h_{ik}(x−w_i)$$
$$\mbox{Si i}   \notin N_k→ w^{new}_i=w^{old}_i$$

El parámetro $\infty$ es la rapidez de aprendizaje, que varia entre 0 y 1, y $h_{ik}$ es un kernel de vecindad y puede ser de diferentes tipos, por ejemplo, gaussiano: 
$$h_{ik} = exp[\frac{-d(r_i - r_k)}{2\sigma²}]$$
Donde $r_i$ y $r_k$ son las posiciones de las celdas i y k en el grupo y $\sigma$ es el radio de la vecindad.

Este paso del algoritmo es el que determina que los datos en dos grupos vecinos sean más similares entre sí que con los datos presentes en grupos más lejanos. El kernel gaussiano determina que la magnitud de los cambios en los pesos baje a medida que las grillas del vecindario están más alejadas de la que recibió el nuevo objeto.

6) Disminuir la rapidez de aprendizaje.

7) Repetir los pasos 3-6 hasta que los pesos $(w)$ se estabilicen.

* De los gráficos se puede decir lo siguiente: 
    1) el grupo uno(azul izquierda) tiene una mayor concentración de agua y lactosa. Pero tiene menor concentración de ash y de proteína. 
    2) el grupo dos(azul derecha) tiene concentración de agua y lactosa, pero no tiene de las demás variables. 
    3) el grupo tres (rojo) tiene una alta concentración de lactosa y de agua, además de una baja concentración de ash.
    4) el grupo cuatro(verde) tiene una alta concentración de lactosa y de ash, pero baja en agua,grasa y proteína.

```r
#create SOM grid
sommap <- som(scale(datos), grid = somgrid(4, 4, "rectangular"))
plot(sommap)

colors <- function(n, alpha = 1) {
    rev(heat.colors(n, alpha))
}

plot(sommap, type = "counts", palette.name = colors, heatkey = TRUE)

plot(sommap, type = "mapping", pchs = 20, main = "Mapping Type SOM")

plot(sommap, type = "dist.neighbours", palette.name = terrain.colors)
```

```r
mat <- matrix(sommap$codes[[1]], ncol=4,nrow=4,byrow=FALSE)

## use hierarchical clustering to cluster the codebook vectors
groups<-3
som.hc <- cutree(hclust(dist(mat)), groups)

#plot
plot(sommap, type="codes", bgcol=rainbow(groups)[som.hc])

#cluster boundaries
add.cluster.boundaries(sommap, som.hc)
```

### Comentarios

* En la mayoría predomina altas concentraciones de agua, seguida de altas concentraciones de lactosa y proteína. La concentración que menos se ve es la de ash y grasa.
