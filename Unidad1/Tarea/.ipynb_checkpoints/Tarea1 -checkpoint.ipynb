**Filtración de variables**

En este caso voy a descartar todas las variables que contengan Id's o que para cada elemento no varien, como: **objid, run, rerun, specobjid, mjd, fiberid**. 

Dejando así algunas variables con Id's como: **camcol, field,plate**, las cuales aunque sea Id's, varian para cada uno de los astros, entonces pienso que pueden otorgar información para las predicciones, no tan útil, pero algo ayudan.

Por otra parte están las variables **ra, dec, u , g , r , i , z , redshift** las cuales si son relevantes e imprecindibles en este trabajo, ya que, las intesidades de brillo son las cuales nos ayudan a identificar los astros en la vida cotidiana y tambien a los astronomicos con los telescopios.

Al final dejamos la variable **class**, la cual nos dice que tipo de astro es, entonces esta variable será nuestra variable independiente(Y) en todos los modelos que trabajaremos más adelante.


```python
##Actualizando data set luego de definir cuales variables se usarán
dataset = df.drop(['objid','run','rerun','specobjid','mjd','fiberid'],1)
dataset.head()
```

# Gráfico que muestra la cantidad de cada astro

```python
sb.countplot(x='class',data=dataset,palette='hls')
```

**Podemos notar que existe una gran diferencia entre la cantidad de QSO's y STAR's - GALAXY's, lo cual es un problema para  realizar unas predicciones más acertadas**

```python
#Cambiando el tipo class por numeros (Star = 0; Galaxy = 1;QSO = 2)
dataset["class"].replace({'STAR':0,'GALAXY':1,'QSO':2},inplace=True)
```

**Antes de realizar las predicciones usando distintos modelos, definiremos algunos conceptos:** 
* Recall : es la fracción de instancias relevantes que se han obtenido sobre la cantidad total de instancias relevantes
* Precision : es la fracción de todas las instancias relevantes dividido entre las instancias obtenidas. 


## Usando el modelo de Árbol de decision 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from IPython.display import Image  
from sklearn import tree
import pydotplus
```

```python
# Variable objetivo
Y = dataset["class"] 
# Variables predictivas
X = dataset[['ra','dec','u','g','r','i','z','field','camcol','redshift','plate']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = .3, random_state=25)

#Entrenamiento
clf =DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
#Predicciones
y_pred = clf.predict(X_test)
#Evaluación del rendimiento del clasificador
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
#Print de la matriz de confusión
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

**Interpretando los resultados de la matrix de confusion:**
* Las **estrellas** tiene un 100% para decir que es una estrella y de todas las estrellas que habían en los datos reconocio un 99% de ellas.
* Las **galaxias** tiene un 97% para decir que es una galaxia y de todas las galaxias que habían en los datos reconocio un 99% de ellas.
* Las **Qsar** tiene un 95% para decir que es una Qsar y de todas las estrellas que habían en los datos reconocio un 90% de ellas.

```python
print('Precisión del modelo:')
print(clf.score(X_train, y_train))
```

```python
features=['ra','dec','u','g','r','i','z','field','camcol','redshift','plate']
classes=['0','1','2']

dot_data = tree.export_graphviz(clf,out_file=None,feature_names=features,class_names=classes, filled=True,rounded=True, impurity=False)

graph = pydotplus.graph_from_dot_data(dot_data) 
s=Image(graph.create_png())
graph.write_png("modelo.png")
s
```

**Conclusiones para Árbol de decisión:**
* Dado el arbol entregado por el modelo, se puede notar que el factor más relevante es **redshift**, ya que, es el primer criterio de separación en el arbol.
* Los siguientes parametro más relevante es **u y g**, al ser hijos de la raíz.
* Así sucesivamente vamos recorriendo el árbol para ver la importancia de cada parametro.

## Usando el modelo de Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=15)
clf = clf.fit(X_train, y_train)

#Predicciones
y_pred = clf.predict(X_test)
#Evaluación del rendimiento del clasificador
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
#Print de la matriz de confusión
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

**Interpretando los resultados de la matrix de confusion:**

* Las **galaxias** tiene un 98% para decir que es una galaxia y de todas las galaxias que habían en los datos reconocio un 99% de ellas.
* Las **Qsar** tiene un 97% para decir que es una Qsar y de todas las estrellas que habían en los datos reconocio un 95% de ellas.

```python
print('Precisión del modelo:')
print(clf.score(X_train, y_train))
```

## Modelo Ordinary Least Squares(OLS)

```python
Y = dataset["class"]
X = dataset[['ra','dec','u','g','r','i','z','camcol','field','redshift','plate']]
scale = StandardScaler()
X_scaled = scale.fit_transform(X[['ra','dec','u','g','r','i','z','camcol','field','redshift','plate']].to_numpy())
#dividimos en sets de entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.75, random_state = 25)
#equilibramos los datos
os = RandomOverSampler()
X_train_res, y_train_res = os.fit_sample(X_train, y_train)
```

```python
import statsmodels.api as sm
est = sm.OLS(y_train_res,X_train_res).fit()
print(est.summary())
```

**Ahora crearemos la función a partir de los coefs que entrega el modelo, esta función nos ayudará a predecir Y(class). Dicha función debería servir, ya que tiene un R-Squared de 0.913, lo cual es bastante aceptable para este problema.**

```python
def predict_iluminati(ra,dec,u,g,r,i,z,camcol,field,redshift,plate):
    return 1.752e-05 *ra +0.0017*dec -0.7743*u+ 1.3898 *g +0.3286*r-0.1089 *i -0.7730*z+0.0038 *camcol-0.0001 *field+0.5712*redshift-9.27e-05*plate
    
```
